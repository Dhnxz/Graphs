```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('9. Sales-Data-Analysis.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

**Bar chart**

Product vs Quantity

```python
product_sales = df.groupby('Product')['Quantity'].sum()

product_sales.plot(kind='bar')

plt.title("Product Sales")
plt.xlabel("Product")
plt.ylabel("Total Quantity Sold")

plt.show()
```

**Pie Chart**

Purchase Type Distribution

```python
df['Purchase Type'].value_counts().plot(
    kind='pie',
    autopct='%1.1f%%'
)

plt.title("Purchase Type Distribution")

plt.ylabel("")

plt.show()
```

**Histogram**

Price Distribution

```python
df['Price'].plot(kind='hist')

plt.title("Price Distribution")

plt.xlabel("Price")

plt.show()
```

**Scatter plot**

Price vs Quantity

```python
plt.scatter(df['Price'], df['Quantity'])

plt.xlabel("Price")
plt.ylabel("Quantity")

plt.title("Price vs Quantity")

plt.show()
```

**Count Plot**

Payment Methods

```python
sns.countplot(x='Payment Method', data=df)

plt.title("Payment Methods")

plt.show()
```

City Sales Analysis

```python
city_sales = df.groupby('City')['Quantity'].sum()

city_sales.plot(kind='bar')

plt.title("City-wise Sales")
plt.xlabel("City")
plt.ylabel("Quantity Sold")

plt.show()
```
